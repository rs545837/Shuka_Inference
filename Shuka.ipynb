```python
# Import required libraries
import os
from pydub import AudioSegment
import librosa
from IPython.display import Audio, display

# Download the sample audio file
!wget -q https://huggingface.co/sarvamai/shuka_v1/resolve/main/hi-question.webm

# Display the original WebM audio
print("Original WebM audio:")
display(Audio("hi-question.webm"))

# Convert WebM to WAV
audio = AudioSegment.from_file("./hi-question.webm", format="webm")
audio.export("./hi-question.wav", format="wav")

# Now load with librosa
audio, sr = librosa.load("./hi-question.wav", sr=16000)

# Your existing code for processing with the pipeline
turns = [
    {'role': 'system', 'content': 'Respond naturally and informatively.'},
    {'role': 'user', 'content': '<|audio|>'}
]

# Assuming 'pipe' is your pipeline function
result = pipe({'audio': audio, 'turns': turns, 'sampling_rate': sr}, max_new_tokens=512)

# Print the result
print("\nPipeline result:")
print(result)
```
